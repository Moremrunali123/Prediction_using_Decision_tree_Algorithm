Import packages

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.tree import plot_tree
from sklearn import tree
```

Dataset

```python
data = pd.read_csv('/content/drive/MyDrive/archive/Iris.csv')
```

```python
from pandas.core.frame import DataFrame
#import dataset
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target']= data.target
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.shape
```

```python
print(df['target'])
```

```python
fc = [x for x in df.columns if x!="target"]
x = df[fc]
y = df["target"]
X_train, X_test, Y_train, Y_test = train_test_split(x,y, random_state = 100, test_size = 0.30)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
model1 = DecisionTreeClassifier()
```

```python
model1.fit(X_train, Y_train)
```

```python
Y_pred = model1.predict(X_test)
```

```python
data2 = pd.DataFrame({"Actual":Y_test,"Predicted":Y_pred})
data2.head()
```

```python
accuracy_score(Y_test,Y_pred)
```

```python
f_n = ["Sepal length", "sepal width", "Petal length", "Petal width"]
c_n = ["Setosa", "Versicolor", "Virginica"]
plot_tree(model1, feature_names = f_n, class_names= c_n , filled = True)
```
